```python
#readline() --> read line by line

f = open('sample_f2.txt','r')
f.readline()
print(f.readline(),end='')
f.close()
f = open('sample_f2.txt','r')
print(f.readline(),end='')
f.close()
```

```python
#reading entire using reading

f = open('sample_f2.txt','r')

while True:
    line = f.readline()

    if line == '':
        break
    else:
        print(line,end='')

f.close()
```

# using Context Manger( with )

    # 1 it's a good idea to close a file after usage as it will free up the resource
    # 2  if we don't close it,garbager collector would close it
    # 3  with keyword closes the file as soon as the usage is over

```python
# with

with open('sample_f2.txt','w') as f:
    f.write('with --> context Manger ')
```

```python
#try to read() on context Manger

with open('sample_f2.txt','r') as f:
    print(f.read())
```

```python
#movinf within a file --> 10 char then 10 char

with open('sample_f2.txt','r') as f:
    print(f.read(10))  # First 10 char print
    print(f.read(10))  #Next 10 charater print
```

```python
# Benefit ? --> to load a big file in memory

big_l = ['hello Rahul cool  ' for i in range(75)]

with open('Big_data_storage.txt', 'w') as f:
    f.writelines(big_l)

```

```python
# we load the file chunk by chuck

with open('big_data_storage.txt', 'r') as f:

    chuck_size = 100

    while len(f.read(chuck_size)) > 0:
        print(f.read(chuck_size), end = '')
        f.read(chuck_size)
```
