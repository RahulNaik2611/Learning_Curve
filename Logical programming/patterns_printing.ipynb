```python
#Simple Star Square

num = 4

for i in range(num):
    for j in range (num):
        print("*" , end= " ")
    print()
```

```python
#right angle triangle

num = 5

for i in range(num):
    print("* " *i)
```

```python
#inverted triangle

num = 5

for i in range(num,0,-1):
    print("* " *i)
```

```python
#left angle triangle

num = 5

for i in range(1,num+1):
    print(" "*(num-1) +" *"*i)
```

```python
#pyramid

num = 5

for i in range(1,num+1):
    print(" "*(num-i)+"* "*i)
```

```python
#inverted pyramid

num = 7

for i in range(num+1, 0 , -1):
    print(" "*(num-i)+ " *"*i)
```

```python
#Diamond

num = 4

for i in range(1,num+1):
    print(" "*(num-i)+ " *"*i)
for i in range(num,0,-1):
    print(" "*(num-i)+ " *"*i)
```

```python
#Hollow square

num = 5

for i in range(num):
    for j in range (num):
        if i == 0 or i == num-1 or j == 0 or j == num-1:
            print("*" , end= " ")
        else:
            print(" ",end=" ")
    print()
```

```python
#Hollow triangle

num = 7

for  i in range (1,num+1):
    for j in range(1,i+1):
        if j == 1 or j == i or i ==num:
            print("*" , end= " ")
        else:
            print(" ",end=" ")
    print()

```

```python
#Number patterns
#Increasing Number

num = 4

for i in range(1,num+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#Same Number Rows

num =7

for i in range(1,num+1):
    for j in range(1,i+1):
        print(i,end=" ")
    print()
```

```python
#Reverse Number

num = 4

for i in range(1,num+1):
    for j in range(i,0,-1):
        print(j,end=" ")
    print()
```

```python
#Floyd's Triangle

num =1

for i in range(1, 5):       # 4 rows
    for j in range(i):      # print 'i' numbers in each row
        print(num, end=" ")
        num += 1
    print()
```
