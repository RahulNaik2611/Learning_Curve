# Speed

# list

```python
a = [i for i in range(10000000)]

b = [i for i in range(10000000,20000000)]

c =[]

import time

Start = time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])

print(time.time() - Start)
```

Checking time with **Numpy**

```python
import numpy as np

a = np.arange(10000000)

b = np.arange(10000000,20000000)

start = time.time()
c = a + b

print(time.time() - start)
```

# Boolean indexing

```python
import numpy as np

twD = np.random.randint(1,100,24).reshape(6,4)
print(twD)
```

```python
#Find the numbers greater than 50

print(twD[twD > 50])
```

```python
#Find out the Even numbers

print(twD[twD % 2 ==0 ])
```

```python
#Find out the all even numbers

print(twD[twD % 2 ==1 ])
```

```python
#Find all number greater than 50 and are even

print(twD[(twD > 50 ) & (twD % 2 ==0 )])
```

```python
#Find all numbers not divisible by 7

print(twD[twD % 7 !=0 ])
```

# **BroadCasting**

1. The term Broadcasting describe how numpy treat arrays with different shapes during arithmetin operations
2. The  smallest array is 'Boardcasting' across the larger array so that they have compatible shape


```python
#Same shape

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print('------------------------------')
print(b)
print('------------------------------')

print(a + b)
```

# Broad Casting rules

1. if the numbers of dimensions of the arrays are different,add new dimension with size 1 to the head of the array with the smaller dimension

## Make each dimension of the two arrays the same size
1. if the sizes of each dimension of the two arrays do not match,dimension with size 1 are stretched to the size of the other array.
2. if there is a dimension whose size is not 1 in either of the two arrays,it cannot be broadcasted,and an error is raised



```python
#More example
import numpy as np
a = np.arange(12).reshape(4,3)

b = np.arange(3)

print(a)

print(b)

print(a+b)
```

# **#Sigmoid**

The sigmoid function always produces values between 0 and 1, and its graph has a smooth S-shape: for large negative inputs it approaches 0, while for large positive inputs it approaches 1, with a gradual transition in between.



```python
#Sigmoid

#Sigmoid function formula  s(X) = 1/1+ e^-x


import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example values
print(sigmoid(-5))  # ~0.0067
print(sigmoid(0))   # 0.5
print(sigmoid(5))
print(sigmoid(np.arange(10)))
```
