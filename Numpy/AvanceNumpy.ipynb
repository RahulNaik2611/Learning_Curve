# Speed

# list

```python
a = [i for i in range(10000000)]

b = [i for i in range(10000000,20000000)]

c =[]

import time

Start = time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])

print(time.time() - Start)
```

Checking time with **Numpy**

```python
import numpy as np

a = np.arange(10000000)

b = np.arange(10000000,20000000)

start = time.time()
c = a + b

print(time.time() - start)
```

# Memory

```python
a = [i for i in range(10000000)]

import sys
print(sys.getsizeof(a))
```

Checking in numpy

```python
import numpy as np
a = np.arange(10000000,dtype=np.int8)
sys.getsizeof(a)
```

Convenience

# Advance indexing



```python
a = np.arange(12).reshape(4,3)
print(a)
```

```python
#Fancy indexing
#Fetch 1,3,4 rows

print(a[[0,2,3]])
```

```python
b = np.arange(24).reshape(6,4)
print(b)

print('------------------------------Fetch rows')
#Fetch the 1,3,4,6 rows

print(b[[0,2,3,5]])

print('------------------------------Fetch column')

print(b[:,[0,2,3]])
```
