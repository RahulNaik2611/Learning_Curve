# Max :

```python
import numpy as np

b = np.arange(12).reshape(3,4)
print(b)

print('---------------------------------------------------------')
# Find the max value in each row (axis=1 → row-wise)
print(f'max value row-wise {np.max(b, axis=1)}')

print('---------------------------------------------------------')
# Find the max value in each column (axis=0 → column-wise)
print(f'max value column-wise {np.max(b, axis=0)}')
```

# min value row & column wise

```python
import numpy as np

b = np.arange(12).reshape(3,4)
print(b)

print('---------------------------------------------------------')
# Find the min value in each row (axis=1 → row-wise)
print(f'min value row-wise {np.min(b, axis=1)}')

print('---------------------------------------------------------')
# Find the min value in each column (axis=0 → column-wise)
print(f'min value column-wise {np.min(b, axis=0)}')
```

# **Sum / Prod Function**

**axis=0 → operate down columns.

axis=1 → operate across rows.**

```python
a = np.arange(12).reshape(3,4)


print(a)
print('-------------------------------------------------------------')
print(f'Total matrix sum  value : {np.sum(a)}')
print('-------------------------------------------------------------')
print(f'Total sum row-wise {np.sum(a, axis=1)}')
print('-------------------------------------------------------------')
print(f'Total sum column-wise {np.sum(a, axis=0)}')
```

prod :

```python
a = np.arange(12).reshape(3,4)


print(a)
print('-------------------------------------------------------------')
print(f'Total matrix prod  value : {np.prod(a)}')
print('-------------------------------------------------------------')
print(f'Total prod row-wise {np.prod(a, axis=1)}')
print('-------------------------------------------------------------')
print(f'Total prod column-wise {np.prod(a, axis=0)}')
```

# mean/median/srd/var

```python
import numpy as np

a = np.arange(12).reshape(3,4)

print(a)
print('-------------------------------------------------------------')
print(f'Total matrix mean value : {np.mean(a)}')
print('-------------------------------------------------------------')
print(f'Total mean row-wise {np.mean(a, axis=1)}')
print('-------------------------------------------------------------')
print(f'Total mean column-wise {np.mean(a, axis=0)}')
print('-------------------------------------------------------------')
print(f'Total matrix median value : {np.median(a)}')
print('-------------------------------------------------------------')
print(f'Total median row-wise {np.median(a, axis=1)}')
print('-------------------------------------------------------------')
print(f'Total median column-wise {np.median(a, axis=0)}')
print('-------------------------------------------------------------')
print(f'Total matrix standard deviation : {np.std(a)}')
print('-------------------------------------------------------------')
print(f'Total std row-wise {np.std(a, axis=1)}')
print('-------------------------------------------------------------')
print(f'Total std column-wise {np.std(a, axis=0)}')
print('-------------------------------------------------------------')
print(f'Total matrix variance : {np.var(a)}')
print('-------------------------------------------------------------')
print(f'Total var row-wise {np.var(a, axis=1)}')
print('-------------------------------------------------------------')
print(f'Total var column-wise {np.var(a, axis=0)}')
```
