## pandas series

```python
#Strings

import numpy as np
import pandas as pd

Counters = ['india','pakistan','usa','srilanka']

print(pd.Series(Counters))
```

```python
marks = [93,63,76,98]
subject = ['science','maths','Telugu','Social']

print(pd.Series(marks,index=subject,name='Mahesh_markList'))
```

Series from Dict

```python
marks = {
    'maths':93,
    'science':63,
    'Telugu':76,
    'Social':98
}

marks_Series = pd.Series(marks,name='Mahesh_markList')
print(marks_Series)
```
