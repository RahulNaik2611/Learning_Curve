```python
import numpy as np
import pandas as pd
import yfinance as yf
import  matplotlib.pyplot as plt
```

```python
#Define the stock to analyze
stocks = ['AAPL','GOOGL','AMZN']

#Define the time period (Last 2 years )

start_date = '2024-01-01'
end_date = '2025-01-01'


#Dowmload data for the each stock
stock_date = yf.download(stocks, start_date, end_date)
```
