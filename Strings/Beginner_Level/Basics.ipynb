```python
# Reverse String

Greeting = "Hello"
print(Greeting[::-1])
```

```python
##Check if a String is palindrome

check = input("Enter a String")

if check == check[::-1]:
    print("it a palindrome")
else:
    print("not a palindrome")
```

```python
#Count the number of vowels in the String

name = input("Enter a name to check the vowels")
vowels = ['a','e','i','o','u','A','E','I','O','U']
count = 0

for ch in name:
    if ch in vowels:
        count += 1


if count > 0:
    print("Number of vowels in the given string:", count)
else:
    print("Vowels not found in given string")


```

```python
#Count number of characters in a string(With out using built-in functions )

name ="Rahul"

count = 0

for ch in name:
    count += 1
print(f'number of characters in the given string: {count}')
```

```python
#Remove the Spaces from a string

sentences = "hi bro how are you"
for ch in sentences:
    if ch != " ":
        print(ch,end="")
    else:
        continue
```
