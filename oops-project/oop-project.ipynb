```python
import nlpcloud
import json


class NLPApp:

    def __init__(self):

        # Replace with your actual NLP Cloud API key
        self.api_key = "a128ede5ea087ccd2ea9f30f8e0001abedb83cd0"

        self.__load_data()

    # =========================
    # Load Users Data
    # =========================
    def __load_data(self):

        try:
            with open("users.json", "r") as f:
                self.__database = json.load(f)

        except FileNotFoundError:
            self.__database = {}

    # =========================
    # Save Users Data
    # =========================
    def __save_data(self):

        with open("users.json", "w") as f:
            json.dump(self.__database, f, indent=4)

    # =========================
    # Main Program
    # =========================
    def run(self):

        while True:

            print("\n===== NLP Application =====")
            choice = input("1.Register\n2.Login\n3.Exit\nEnter choice: ")

            if choice == '1':
                self.__register()

            elif choice == '2':

                result = self.__login()

                if result:
                    self.__menu()

            elif choice == '3':
                print("Application Closed")
                break

            else:
                print("Invalid Choice")

    # =========================
    # Register User
    # =========================
    def __register(self):

        email = input("Enter Email: ")

        if email in self.__database:
            print("User already exists")
            return

        name = input("Enter Name: ")
        password = input("Enter Password: ")

        self.__database[email] = {
            "name": name,
            "password": password
        }

        self.__save_data()

        print("Registration Successful")

    # =========================
    # Login User
    # =========================
    def __login(self):

        email = input("Enter Email: ")
        password = input("Enter Password: ")

        if email in self.__database:

            if self.__database[email]["password"] == password:

                print(f"Welcome {self.__database[email]['name']}")
                return True

        print("Invalid Email or Password")
        return False

    # =========================
    # Menu After Login
    # =========================
    def __menu(self):

        while True:

            print("\n===== NLP Features =====")

            choice = input(
                "1.Named Entity Recognition\n"
                "2.Language Detection\n"
                "3.Sentiment Analysis\n"
                "4.Logout\n"
                "Enter choice: "
            )

            if choice == '1':
                self.__ner()

            elif choice == '2':
                self.__language_detection()

            elif choice == '3':
                self.__sentiment()

            elif choice == '4':
                print("Logged Out")
                break

            else:
                print("Invalid Choice")

    # =========================
    # Named Entity Recognition
    # =========================
    def __ner(self):

        para = input("Enter Text: ")
        search = input("Enter Entity Type (person/org/location): ")

        client = nlpcloud.Client(
            "finetuned-llama-3-70b",
            self.api_key,
            gpu=True
        )

        response = client.entities(
            para,
            searched_entity=search
        )

        print("\nNER Result:")
        print(response)

    # =========================
    # Language Detection
    # =========================
    def __language_detection(self):

        para = input("Enter Text: ")

        client = nlpcloud.Client(
            "langdetect",
            self.api_key
        )

        response = client.langdetection(para)

        print("\nLanguage Detection Result:")
        print(response)

    # =========================
    # Sentiment Analysis
    # =========================
    def __sentiment(self):

        para = input("Enter Text: ")

        client = nlpcloud.Client(
            "distilbert-base-uncased-finetuned-sst-2-english",
            self.api_key
        )

        response = client.sentiment(para)

        print("\nSentiment Result:")
        print(response)

    # =========================
    # Start Application
    # =========================


if __name__ == "__main__":

    obj = NLPApp()
    obj.run()
```
